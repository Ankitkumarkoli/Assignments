# Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

```python
# LookupError class is used when an index or a key is not found for sequence or dictionary.
# LookupError exception class handle both the KeyError and IndexError.
```

```python
# IndexError
import logging 

logging.basicConfig(filename = "king.log" , level = logging.ERROR)

a = [1,2,3,4,5]

try:
    result = a[10]
    
except IndexError as e:
    logging.error("I am handling the indexerror{}".format(result))
    
```

```python
# KeyError

emp_dict = {'Name': 'Pankaj', 'ID': 1}

try:
    emp_id = emp_dict['ID']
    print(emp_id)

    emp_role = emp_dict['Role']
    print(emp_role)
except KeyError as ke:
    print('Key Not Found in Employee Dictionary:', ke)
```

```python
#KeyError
a = {"name":"ankit","course":"Datascience_masters"}

try:
    s = a["language"]
    print(s)
    
except KeyError as e:
    print(e)
    
    
```
